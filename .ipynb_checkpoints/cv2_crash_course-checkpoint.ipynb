### Introduction to Open  cv

Here in this notebook, we are looking into the basics of open cv such as reading an image, shape of an image, converting an image into grayscale. <br>
Basic cv2 functions used includes;<br><br>
`imread()` => to read an image and convert it into arrays <br>
`imshow()` => displayes an image <br>
`waitKey()` => delay time an image is meant to be displayed <br>
`destroyAllWindows()` => closes the display image <br>
`imwrite()` => save an image <br>
`cv2.cvtColor(img, cv2.COLOR_BGR2RGB)` => convert images to grayscale

```python
# import the library 
import cv2
import numpy as np 
```

```python
# loading an image using "imread" function and passing the image path as an argument
image = cv2.imread("wayne.jpg")
```

```python
# checking the dimension of the image which are represented as (Height, width, channel)
print(image.shape)
```

```python
# show the loaded image
cv2.imshow('first image', image)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# we can make the image to be grayscale by passing the value "0" in imread function. 
image = cv2.imread("wayne.jpg", 0)
cv2.imshow("black and white", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# arrays of black and white images and it shape . As we can see, the gray scale image is in 2-D array
image.shape
```

```python
# alternative way of making an image greyscale
image_data = cv2.imread('wayne.jpg')
grayscale_image = cv2.cvtColor(image_data, cv2.COLOR_BGR2GRAY)
cv2.imshow('alt_gray', grayscale_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# saving the image. we use the function "imwrite" that takes in two parameters
# 1. the name we want to save the image as. include the image format (.jpg, .png, etc) 
# 2. the image variable we are saving.    Now saving the grayscaled image

cv2.imwrite('gary_wayne.jpg', grayscale_image)
```
